```python
import pandas as pd 
import matplotlib.pyplot as mt
import seaborn as sns 
```

```python
df = pd.read_csv("Airbnb_NYC_2019.csv")

df.head(5)
```

```python
df.shape
```

```python
# 10 A
manhattan_neighouring_group = df[df['neighbourhood_group'] == 'Manhattan']
manhattan_neighouring_group.shape
```

```python
manhattan_neighouring_group
```

```python
mt.figure(figsize= (20,9))
sns.countplot(x = 'price', data = manhattan_neighouring_group)
mt.title('Distribution of prices for the Manhattan listings. ')
mt.xticks(rotation=45, horizontalalignment='right')
mt.show()
```

```python
#create histogram to visualize:
# a. histrogram
mt.figure(figsize=(40,18))
sns.countplot(x = 'price', data = manhattan_neighouring_group)
mt.title('Distrubition of prices for Manhattan in 2019')
mt.show()
```
